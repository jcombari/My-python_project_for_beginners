```python
speed(0)
screen = Screen()
colormode(255)

from math import*
import random

import os
os.chdir("Your Directory")

def create_rhombus(sz,stx,sty,col,rt_col):
   up()
   goto(stx,sty)
   down()
   if rt_col==1:
       fillcolor(80,119,117)
   if col=="yes":
       fillcolor(50,25,0)
   begin_fill()
   for  i in range(2): 
       fd(sz)
       left(60)
       fd(sz)
       left(120)
   end_fill()
   fillcolor(192,192,192)    
       
   fd(sz)
   right(60)
   
   if rt_col==2:
       fillcolor(50,25,0)
   if col=="yes":
       fillcolor(80,119,117)    
   begin_fill()
   for  i in range(2): 
       fd(sz)
       left(120)
       fd(sz)
       left(60)
   end_fill() 
   fillcolor(192,192,192)   
       
   left(120)
   fd(sz)
   left(120)
   fd(sz)
   right(60)
   fd(sz)
   right(120)
   


def gen_pattern(sz,mul):
    yv = int(1200/(2*sz*sin(pi/3)))
    xv = int(1200/(2*sz))
    cnt=1       
    for y in range(yv):
        sty = -500 + y*(2*sz*sin(pi/3))
        for x in range(xv):
            stx = -700 + x*2*sz 
            if (cnt-2)%mul==0:
                col="yes"
            else:
                col="no"          
            r_col = random.randint(0,100)
            rt_col=0
            if r_col%9==0:
                 rt_col = random.randint(1,2)
            create_rhombus(sz,stx,sty,col,rt_col)
            cnt+=1
       
def gen_bg():
    up()
    goto(-700,-700)
    down()
    fillcolor(192,192,192)
    begin_fill()
    for i in range(4):
        fd(1400)
        left(90)
    end_fill()    
            
tracer(0,0)
gen_bg()
gen_pattern(30,6)
ht()
#create_rhombus(30,0,0)
update() 
screen.getcanvas().postscript(file='Pavement_Tiles.eps') 
done() 
  
```
